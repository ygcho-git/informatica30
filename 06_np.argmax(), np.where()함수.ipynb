#np.max() vs np.argmax()

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4, 5])

print(np.max(arr1))     # np.max() : 최대값
print(np.argmax(arr1))  # np.argmax() : 최대값의 인덱스 번호
```

```python
# 2차원 배열에서
arr2 = np.array([[2, 7, 8, 6, 3],
                 [3, 2, 1, 9, 0],
                 [6, 3, 1, 2, 5]])

print(np.max(arr2))
print(np.argmax(arr2))
```

```python
# 각 1차원 배열에서 최대값, 최대값의 index번호 ?
arr2 = np.array([[2, 7, 8, 6, 3],
                 [3, 2, 1, 9, 0],
                 [6, 3, 1, 2, 5]])

print(np.max(arr2))
print(np.argmax(arr2))
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 2, 7, 3, 5])

result1 = arr1>3
result2 = arr2>=5

print(result1)
print(result2)
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 2, 7, 3, 5])

result = arr1 == arr2

print(result)
```

# np.where() # True인 값의 인덱스 번호를 numpy 배열로 리턴

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 2, 7, 3, 5])

result = np.where(arr1==arr2)

print(result)
```
