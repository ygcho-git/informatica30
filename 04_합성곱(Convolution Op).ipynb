```python
from scipy import misc
import numpy as np

Edge1 = np.array([[0,  1, 0],
                  [1, -4, 1],
                  [0,  1, 0]])

Edge2 = np.array([[-1, -1, -1],
                  [-1,  8, -1],
                  [-1, -1, -1]])

Sharpen = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])

Blur = np.array([[1/16, 2/16, 1/16],
                 [2/16, 4/16, 2/16],
                 [1/16, 2/16, 1/16]])

image = misc.face()

Feature_map = np.zeros((766, 1022, 3))

# 이미지 크기 : (768, 1024, 3)  높이 : 768, 너비 : 1024, 채널 : 3(RGB)
for i in range(0, 768, 1) :
    for j in range(0, 1024, 1) :
        Feature_map[i, j, 0] = np.sum(image[i:i+3, j:j+3, 0] * Edge1)  # R채널
        Feature_map[i, j, 1] = np.sum(image[i:i+3, j:j+3, 1] * Edge1)  # G채널
        Feature_map[i, j, 2] = np.sum(image[i:i+3, j:j+3, 2] * Edge1)  # B채널

plt.imshow(Feature_map)
```
