# 결측치 처리

image.png

```python
import numpy as np

# 다시 원래 데이터로
L = [["손흥민", "축구", "남", 30],
     ["김민재", "축구", "남", np.nan],
     ["김연경", "배구", "여", 35],
     ["박찬호", "TMI", "남", np.nan],
     ["지소현", np.nan, "여", 32]]

df = pd.DataFrame(L, columns=["성명", "종목", "성별", "나이"])

df
```

```python
df.info()
# 데이터 프레임에 대한 정보 확인하기
```

```python
# 각 항목별 결측치 확인하기
df.isnull().sum()
```

```python
# 결측치를 임의의 원하는 값으로 채우기

df.loc[:, "나이"] = df["나이"].fillna(0)  # 0으로 채우기

# 나이의 평균 : df["나이"].mean()
df
```

```python
# 결측치가 있는 행 제거

df.dropna(axis=0)
```
